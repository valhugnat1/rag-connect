```python
import pandas as pd
import numpy as np
import re
import nltk
import spacy
import string
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./data/sample.csv')
df.head(n= 50)
```

```python
def display_tweets_recursive(tweet_id, df):
    # Find the rows with the given tweet_id
    rows = df[df['tweet_id'] == tweet_id]

    # Check if any rows are found
    if not rows.empty:
        # Use the first row
        tweet_row = rows.iloc[0]

        # Display the tweet text
        text = (
            f'Inbound: {tweet_row["inbound"]}\n'
            f'Tweet ID: {tweet_id}\n'
            f'{tweet_row["author_id"]}: {tweet_row["text"]}\n'
            f'{"-" * 50}\n\n'
        )

        # Check if there is a response tweet
        response_tweet_id = tweet_row['response_tweet_id']
        if pd.notna(response_tweet_id):
            # Recursive call to display the response tweet
            try : 
                response_tweet_id = int(response_tweet_id)
            except : 
                response_tweet_id = 0

            other_rep = display_tweets_recursive(response_tweet_id, df)
            return text + other_rep 
        else  : 
            return text

    return ''

# Example usage: Display tweets for a specific tweet_id
tweet_id_to_display = 119253
print(display_tweets_recursive(tweet_id_to_display, df))
```

```python
# Filter rows with non-null in_response_to_tweet_id and where inbound is True
filtered_df = df[df['in_response_to_tweet_id'].isnull() & df['inbound']]


# Create a file named "tweet_{tweet_id}_response_to_{response_tweet_id}.txt"
file_name = f'tweet_response.txt'
output_file = 'all_tweets.txt'

# Write tweet text to the file
# Iterate over the filtered DataFrame and write tweets to the same file with newline separators
with open(output_file, 'w', encoding='utf-8') as file:
    for index, row in filtered_df.iterrows():
        tweet_text = display_tweets_recursive(row['tweet_id'], df)
        file.write(tweet_text + '\n\n\n')

print("Text files created successfully.")
```
